```python
# Filtering out rows with non-null descriptions
df_filtered = df.dropna(subset=['description'])

print("Percentage of Null Values in 'description' column:", (df['description'].isnull().sum() / len(df['description'])) * 100
)
print("\nFiltered DataFrame with Non-null Descriptions:")
df_filtered
```

```python
df_average_ratings = pd.read_csv('/content/drive/MyDrive/rating prediction/avg_stars_df_run2 (1) (1).csv') #this df contains the average of ratings of each customer
df_average_ratings
```

```python
average_ratings = df_average_ratings.set_index('user_id')['star'].to_dict()
```

```python
# Counting the number of comments per star
df_filtered.groupby('star')['description'].count()
```

```python
from sklearn.utils import resample

# Resampling each rating category to have the same count as the least frequent category
final_df = []
for star in df_filtered['star'].unique():
    star_df = df_filtered[df_filtered['star'] == star]
    final_df.append(resample(star_df, replace=True, n_samples=250, random_state=42))

final_df = pd.concat(final_df)
final_df
```

```python
import torch
from transformers import BertTokenizer, BertModel
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained("bert-base-multilingual-cased")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)
```

```python
def get_embeddings(text_list):
    encoded_inputs = tokenizer(text_list, padding="max_length", truncation=True, return_tensors="pt")
    encoded_inputs = encoded_inputs.to(device)
    with torch.no_grad():
        outputs = model(**encoded_inputs)
    return outputs.pooler_output #sentence embedding from pooler output
```

```python
# Apply the function to get embeddings for all text in final_df
final_df["embeddings"] = final_df["description"].apply(lambda x: get_embeddings([x]))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
import transformers
```

```python
model_path = "google-bert/bert-base-multilingual-cased"

model_kwargs = {'device':'cuda'}

encode_kwargs = {'normalize_embeddings': False}

embeddings = HuggingFaceEmbeddings(
    model_name=model_path,     # Providing the model's path
    model_kwargs=model_kwargs, # model configuration options
    encode_kwargs=encode_kwargs # encoding options
)
```

```python
documents = [Document(page_content=d['description'], metadata={"rating": d['star'],"user_id": d['user_id'],"comment_id":d["_id"],"embedding":d["embeddings"],"id": id}) for id, d in final_df.iterrows()]

db = FAISS.from_documents(documents, embeddings)

retriever = db.as_retriever(search_kwargs={"k":9})
```

```python
#getting a sample of final_df
sample_df = final_df.sample(frac=1).reset_index(drop=True)
```

```python
import networkx as nx
```

```python
G=nx.Graph()
```

```python
#building a graph from the retrieved texts
for text,id,comment_id,star in zip(sample_df["description"],sample_df["user_id"],sample_df["_id"],sample_df["star"]):
    similar_docs = retriever.get_relevant_documents(text)[1:]
    for doc in similar_docs:
        G.add_edge(comment_id,doc.metadata["comment_id"])
        G.nodes[comment_id]["role"] = star
```

```python
# Remove self-loops
G.remove_edges_from([(n, n) for n in G.nodes() if G.has_edge(n, n)])
pos = nx.spring_layout(G)

nx.draw(G, pos, edge_color='lightgray',node_size=10)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
# Defining the roles and corresponding colors
roles = [1, 2, 3, 4, 5]
role_colors = cm.Blues(np.linspace(0.3, 1, len(roles)))

# color mapping
color_map = {role: color for role, color in zip(roles, role_colors)}

# Draw the graph
pos = nx.spring_layout(G)
node_colors = [color_map[G.nodes[node].get("role", "unknown")] for node in G.nodes()]
nx.draw(G, pos, edge_color='lightgray', node_size=10, node_color=node_colors)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from torch_geometric.nn import GATConv
```

```python
#adjacency matrix
node_ids = list(G.nodes())
adjacency_matrix = nx.adjacency_matrix(G,nodelist=node_ids)
adjacency_matrix = adjacency_matrix.toarray()
```

```python
adj_tensor = torch.tensor(adjacency_matrix, dtype=torch.float32)
edge_index = adj_tensor.nonzero().t()
```

```python
# Drop duplicates
sample_df_copy=sample_df.copy()
sample_df_copy= sample_df_copy.drop_duplicates(subset=['_id'], keep='first')
```

```python
target_col = sample_df_copy.loc[sample_df_copy['_id'].isin(node_ids), 'star']

features = sample_df_copy.loc[sample_df_copy['_id'].isin(node_ids), 'embeddings']

embeddings = [tensor[0].tolist() for tensor in features]

node_features = torch.tensor(embeddings, dtype=torch.float32)

labels = torch.tensor(target_col.values, dtype=torch.float)
```

```python
from sklearn.decomposition import PCA

# Creating a PCA object with desired number of components
pca = PCA(n_components=10)

pca.fit(node_features)

# Transform the node features
reduced_features = torch.tensor(pca.transform(node_features))
reduced_features
```

```python
#adding average ratings as feature
userids = sample_df_copy.loc[sample_df_copy['_id'].isin(node_ids), 'user_id']

# Extracting values and convert to tensor
ratings = torch.tensor([float(average_ratings[key]) for key in userids])

# Adding the new column to the other tensor
node_features = torch.cat((reduced_features, ratings.view(len(node_ids), 1)), dim=1)
node_features = torch.tensor(node_features, dtype=torch.float32)
```

```python
node_features.size()
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
data = Data(x=node_features, edge_index=edge_index,y=labels)
data = data.to(device)
data.x.shape
```

```python
class GATRegression(torch.nn.Module):
    def __init__(self, num_node_features):
        super(GATRegression, self).__init__()
        self.conv1 = GATConv(11, 32, heads=4)
        self.conv2 = GATConv(32*4, 32, heads=2)
        self.conv3 = GATConv(32*2, 16)
        self.linear = nn.Linear(16 , 1)


    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.tanh(x)
        x = self.conv2(x, edge_index)
        x = F.tanh(x)
        x = self.conv3(x, edge_index)
        x = F.tanh(x)
        x = self.linear(x.squeeze(1))
        return x
```

```python
num_nodes = data.num_nodes

# Creating a train_mask
train_mask = torch.zeros(num_nodes, dtype=torch.bool)
train_mask[:960] = True
train_mask = train_mask.to(device)

data.train_mask = train_mask

test_mask = torch.logical_not(train_mask)
data.test_mask = test_mask
```

```python
model = GATRegression(11)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

num_epochs=1000

model = model.to(device)

criterion = nn.MSELoss()

model.train()

for epoch in range(num_epochs):
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    # Calculate RMSE
    train_rmse = torch.sqrt(nn.MSELoss()(out[data.train_mask], data.y[data.train_mask]))
    train_absolute_errors = torch.abs(out[data.train_mask] - data.y[data.train_mask])
    train_percentage_errors = train_absolute_errors / data.y[data.train_mask] * 100
    train_mape = torch.mean(train_percentage_errors)

    loss.backward()
    optimizer.step()

    if (epoch+1) % 100 == 0:
        print('Epoch: {},  train Loss: {:.4f}, Training RMSE: {:.4f} , training MAPE :{:.4f}'.format(epoch+1, loss.item(), train_rmse.item(),train_mape.item()))
        with torch.no_grad():
            test_out = model(data.x, data.edge_index)
            loss = criterion(out[data.test_mask], data.y[data.test_mask])
            # Calculate test RMSE
            test_rmse = torch.sqrt(nn.MSELoss()(test_out[data.test_mask], data.y[data.test_mask]))
            test_absolute_errors = torch.abs(test_out[data.train_mask] - data.y[data.train_mask])
            test_percentage_errors = test_absolute_errors / data.y[data.train_mask] * 100
            test_mape = torch.mean(test_percentage_errors)
            print('Epoch: {}, test Loss: {:.4f} Test RMSE: {:.4f},  training MAPE :{:.4f}\n'.format(epoch + 1, loss.item(), test_rmse.item(),test_mape.item()))
```
